```python
import cv2
import numpy as np
import tensorflow as tf


import serial.tools.list_ports

ports=serial.tools.list_ports.comports()
serialInst=serial.Serial()
portList=[]

for onePort in ports:
    portList.append(str(onePort))
    print(str(onePort))

val=input("COM:")
for x in range(0,len(portList)):
    if portList[x].startswith("COM"+str(val)):
        portVar="COM"+str(val)
        print(portVar)

serialInst.baudrate=9600
serialInst.port=portVar
serialInst.open()
default='60'
#serialInst.write(default.encode('utf-8'))
#url="https//192.168.3.85:8080/vod"
# Initialize the mobile camera
#cap = urllib.request.urlopen(url).read()
window_name = 'cocpit-digital'
cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)


png_path = 'compt.png'  # Replace with the actual path to your PNG image file
png_image = cv2.imread(png_path, cv2.IMREAD_UNCHANGED)

#scale_percent = 25  # Adjust the scaling factor as needed
#width = int(png_image.shape[1] * scale_percent / 100)
#height = int(png_image.shape[0] * scale_percent / 100)
png_image = cv2.resize(png_image, (200, 200))

offset_x = 350  # Adjust the x-coordinate offset as needed
offset_y = 100  # Adjust the y-coordinate offset as needed
#Open a video capture object for the camera (0 represents the default camera)
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the camera feed
    ret, frame = cap.read()
    # Check if the frame is valid
    if not ret:
        print("Failed to retrieve frame from the IP webcam stream")
        break
    # Define the ROI coordinates
    x, y, width, height = 100, 100, 200, 200


    # Draw a rectangle to visualize the ROI
    cv2.rectangle(frame, (x, y), (x + width, y + height), (0, 255, 0), 2)
    
    # Extract the ROI from the frame
    roi = frame[y:y + height, x:x + width]
     
    #x2, y2, width2, height2 = 0, 350, 640, 130
    #cv2.rectangle(frame, (x2, y2), (x2 + width2, y2 + height2), (0, 0, 0), -1)
    #roi2 = frame[y:y2 + height2, x:x2 + width2]
    # Overlay the PNG image on the top-left corner of the frame
    frame[offset_y:offset_y+height, offset_x:offset_x+width] = png_image     #haaaaaaaaaaaaaaadaaaaaaaaaaa
    
    

    # Preprocess the frame
    image = cv2.resize(roi, (30, 30))
    image = np.expand_dims(image, axis=0)
    image = np.array(image)
    print(image.shape)
    pred = model.predict([image])[0]
    m=max(pred)
    sign = classes[(np.argmax(pred))+1]
    
    cv2.putText(frame, str(sign) , (425, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    #cv2.putText(frame,'Dernier Panneau :' + str(sign) + ' km/h', (0, 380), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv2.imshow(window_name, frame)
    
    # Display the predicted class label on the frame
    #cv2.putText(frame, str(sign)+" km/h", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv2.imshow('Real-time Classification', frame)
    #write command to arduino 
    if cv2.waitKey(1) & 0xFF==ord(' '):
        serialInst.write(bytes([sign]))
        #cv2.putText(frame, "envoi en cours..", (15, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        break
    else:
        serialInst.write(bytes(0))
    # Break the loop if 'q' is pressed
    #if cv2.waitKey(1) & 0xFF == ord('q'):
        

# Release the video capture object and close the windows
cap.release()
cv2.destroyAllWindows()
```
